# Dataframes

Um DataFrame é uma das estruturas de dados mais importantes e amplamente utilizadas na biblioteca Pandas do Python. Ele é essencialmente uma tabela bidimensional, semelhante a uma planilha do Excel ou uma tabela de banco de dados, onde os dados são organizados em linhas e colunas.

Principais Características de um DataFrame:
1. **Estrutura Bidimensional**: Os dados são organizados em linhas e colunas.

2. **Rotulagem**: Tanto as linhas quanto as colunas podem ser rotuladas, o que facilita o acesso e a manipulação dos dados.

3. **Tipos de Dados Diversos**: Cada coluna pode conter diferentes tipos de dados   (números, strings, booleanos, etc.).
